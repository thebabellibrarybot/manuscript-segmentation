```python
# setup print statements for model eval

def train(train_loader, test_loader, train_ds, test_ds, Net, num_epochs, batch_size, Optimizer, criterion):
    trainLoader = train_loader
    testLoader = test_loader
    
    
    print('[INFO] training the network... ')
    startTime = time.time()
    trainSteps = len(train_ds) // batch_size
    testSteps = len(test_ds) // batch_size
    H = {'train_loss': [], 'test_loss': []}

    # train loop
    for e in tqdm(range(num_epochs)):
        print(e)
        Net.train()
        totaltrainloss = 0
        totaltestloss = 0
        for i, d in enumerate(trainLoader):
            # load to model for pred
            img = d['img_data']
            label = d['label_data']
            label = torch.argmax(label, dim=1) # adjust shape for CEL

            img = torch.autograd.Variable(img, requires_grad = False).to(seg_config.DEVICE)
            label = torch.autograd.Variable(label,requires_grad = False).to(seg_config.DEVICE)
            Pred = Net(img)['out']

            Loss = criterion(Pred, label.long())
            Loss.backward()
            Optimizer.step()

            totaltrainloss += Loss

        # eval loop
        with torch.no_grad():
            Net.eval()
            for itr, d in enumerate(testLoader):
                # load to model for pred
                img, label = d['img_data'], d['label_data']
                label = torch.argmax(label, dim=1) # adjust shape for cel
                img = torch.autograd.Variable(img, requires_grad = False).to(seg_config.DEVICE)
                label = torch.autograd.Variable(label,requires_grad = False).to(seg_config.DEVICE)
                Pred = Net(img)['out']

                Loss = criterion(Pred, label)
                totaltestloss += Loss

        avgtrainLoss = totaltrainloss / trainSteps
        avgtestLoss = totaltestloss / testSteps
        H['train_loss'].append(avgtrainLoss.cpu().detach().numpy())
        H['test_loss'].append(avgtestLoss.cpu().detach().numpy())
        print('[INFO] Epoch: {}/{}'.format(e + 1, seg_config.NUM_EPOCHS))
        print('Train loss: {:.6f}, Test loss {:.4f}'.format(avgtrainLoss, avgtestLoss))
        if itr == 50 or 100 or 150: 
            print('Saving Model.torch')
            torch.save(Net.state_dict(), seg_config.MODEL_PATH)
            plt.style.use("ggplot")
            plt.figure()
            plt.plot(H["train_loss"], label="train_loss")
            plt.plot(H["test_loss"], label="test_loss")
            plt.title("Training Loss on Dataset")
            plt.xlabel("Epoch #")
            plt.ylabel("Loss")
            plt.legend(loc="lower left")
            plt.savefig(seg_config.PLOT_PATH)
        if itr == seg_config.NUM_EPOCHS:
            break

    endTime = time.time()
    print('[INFO] total time taken to train the model: {:.2f}s'.format(endTime - startTime))
```
